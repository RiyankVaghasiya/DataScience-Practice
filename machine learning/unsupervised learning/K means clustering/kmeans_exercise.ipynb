```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
dataset = load_iris()
```

```python
df = pd.DataFrame(data = dataset.data,columns=dataset.feature_names)
```

```python
df.drop(columns=['sepal length (cm)','sepal width (cm)'],index=1,inplace=True)
```

```python
df.head()
```

```python
plt.scatter(data = df , x='petal length (cm)',y='petal width (cm)')
```

__now we have to find that maximum how many cluster we can form__

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,10):
    kf = KMeans(n_clusters= i)
    kf.fit_predict(df)
    wcss.append(kf.inertia_)
    
```

```python
wcss
```

```python
plt.plot(range(1,10), wcss)
```

__now we can see that we can form 3 cluster because our elbow point is 3__

```python
km = KMeans(n_clusters=3)
y_pred = km.fit_predict(df)
```

```python
y_pred
```

```python
df['cluster'] = y_pred
```

```python
df.head()
```

```python
df0 = df[df['cluster'] == 0]
df1 = df[df['cluster'] == 1]
df2 = df[df['cluster'] == 2]
```

```python
plt.scatter(x = df0['petal length (cm)'],y = df0['petal width (cm)'] , marker='+')
plt.scatter(x = df1['petal length (cm)'],y = df1['petal width (cm)'] , marker='^')
plt.scatter(x = df2['petal length (cm)'],y = df2['petal width (cm)'] , marker='*')
```
