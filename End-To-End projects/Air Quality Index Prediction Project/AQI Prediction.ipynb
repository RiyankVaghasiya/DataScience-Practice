```python
date_info = pd.to_datetime(df['Date'],format='%d/%m/%Y')
```

```python
time_info = df['Time'].apply(lambda x:x.replace('.',':'))
```

```python
date_time = pd.concat([date_info,time_info],axis=1)
```

```python
date_time['ds'] = date_time['Date'].astype(str) +' ' + date_time['Time'].astype(str)
```
