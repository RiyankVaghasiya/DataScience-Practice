```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

__scatter plot__

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',marker='+',s=100,hue='sex')
```

__line plot__

```python
gap = px.data.gapminder()
```

```python
gap.sample(5)
```

```python
india = gap[gap['country'] == 'India']
```

```python
sns.lineplot(data = india, x = 'year' , y='lifeExp')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
```

```python
sns.lineplot(data = temp_df,x='year',y='lifeExp',hue='country',style='continent')
```

__facet plot__

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

__distribution plots__

```python
sns.displot(data=tips,x='total_bill',kind='hist',hue='sex')  #we can also use sns.histplot() function to plot a graph
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.displot(data=titanic,x='age',kind='hist',col='sex')
```

```python
#kde plot
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,kind='kde',x='total_bill',hue='sex',fill=True)
```

```python
#heatmap
#we can not plot directly heatmap firstly we need to convert it in pivot table
```

```python
gap_pivot = gap.pivot_table(index = 'country',columns='year',values='lifeExp')
```

```python
gap_pivot.head()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=gap_pivot)
```

```python
europe = gap[gap['continent'] == 'Europe']
```

```python
europe_pivot = europe.pivot_table(index = 'country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data = europe_pivot,cmap='summer',annot=True)
```

__categorical plot__

```python
sns.catplot(data =tips,x='day',y='total_bill',kind='strip')
```

```python
#box plot
sns.catplot(data=tips,kind='box',x='day',y='total_bill',hue='sex')
```

```python
#violin plot
sns.violinplot(data=tips,x = 'sex',y='total_bill',color='pink')  #we can also plot this graph using sns.catplot()
```

```python
#point plot
sns.catplot(data=tips,kind='point',x='sex',y='total_bill',hue='smoker')
```

__regression plots__

```python
sns.regplot(data=tips,x='total_bill',y='tip')
```
