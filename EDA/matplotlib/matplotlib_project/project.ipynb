```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

__Line Graph__

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas.head()
```

```python
plt.figure(dpi=100)
plt.plot(gas.Year,gas.USA ,'b.-',label='USA');
plt.plot(gas.Year,gas.Australia,'r.-',label='Australia')
plt.plot(gas.Year,gas.Germany,'g.-',label='Germany')
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.title('Gas Price Over Time(In USD)')
plt.xticks(gas.Year[::3])
plt.legend()
```
