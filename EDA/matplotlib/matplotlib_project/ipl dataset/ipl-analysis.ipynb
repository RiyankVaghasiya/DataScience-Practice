```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('sharma-kohli.csv')
```

```python
data
```

```python
plt.plot(data['index'],data['RG Sharma'],'m',linestyle='--',marker='o',markersize='8',label='R Sharma')
plt.plot(data['index'],data['V Kohli'],'yellow',linestyle='-',marker='D',markersize='5',label='V Kohli')
plt.xlabel('season')
plt.ylabel('IPL runs')
plt.title('Runs scored by kohli and sharma')
plt.legend(loc='best')
#plt.xlim(2010,2015)
#plt.ylim(200,500)
plt.grid()
plt.show
```

__stacked bar chart__

__annotations__

```python
annotation = pd.read_csv('batter.csv')
```

```python
sample_df = annotation.head(100).sample(25)
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'] , sample_df['strike_rate'],s=sample_df['runs']) #this s param will create bubble according to run size
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i] ,sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.grid()
plt.axhline(130,color='red') #for horizantal line
plt.axvline(30,color='red') #for vertical line
plt.show()
```

__make subplots__

```python
#using subplots we can plot multiple graph simultaneously
fig , ax = plt.subplots(nrows=2,ncols=1)
ax[0].scatter(annotation['avg'], annotation['strike_rate'])
ax[1].scatter(annotation['avg'],annotation['runs'],color='magenta')
ax[0].set_title("average and strike rate comparison")
ax[1].set_title("average and runs comparison")
ax[0].set_xlabel("average of batsman", color='red')
ax[0].set_ylabel("strike rate of batsman" , color='red')
ax[1].set_xlabel("average of batsman", color='red')
ax[1].set_ylabel("runs of batsman", color='red')


plt.subplots_adjust(wspace=0.5, hspace=0.6)
```

```python
fig , ax = plt.subplots(nrows=2,ncols=2)
ax[0,0].scatter(annotation['avg'], annotation['strike_rate'])
ax[0,1].hist(annotation['avg'],color='orange')
ax[1,0].scatter(annotation['avg'], annotation['runs'],color='red')
ax[1,1].hist(annotation['runs'],color='magenta')
plt.show()

```

__3D Plots__

```python
fig = plt.figure(figsize=(15,7))
ax = plt.subplot(projection = '3d')
ax.scatter3D(annotation['avg'], annotation['strike_rate'],annotation['runs'])
ax.set_title('ipl batsman analysis')
ax.set_xlabel('average')
ax.set_ylabel('strike_rate')
ax.set_zlabel('runs')
```

__3D line plot__

```python
x=[0,1,2,25]
y=[0,10,13,0]
z =[0,13,20,9]

fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=50)
ax.plot3D(x,y,z,color='red')
```

__3D surface plot__

```python
x = np.linspace(-5, 5, 50)  
y = np.linspace(-5, 5, 50) 
x, y = np.meshgrid(x, y)  
```

```python
z = np.sin(x)**2 + np.cos(y)**2
fig = plt.figure(figsize=(15,7))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(x,y,z,cmap='viridis')
fig.colorbar(p)
```

__contour plots__

```python
fig = plt.figure(figsize=(10,6))
ax = plt.subplot()
p = ax.contourf(x,y,z,cmap='viridis')
fig.colorbar(p)
```

__heatmap__

```python
heat = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
```

```python
heat.head()
```

```python
# filter data we need only 1 to 6 balls
heat['ballnumber'].unique()
```

```python
mask = (heat['ballnumber'] <= 6) & (heat['batsman_run'] == 6)
heat = heat[mask]
```

```python
grid =heat.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(15,7))
plt.imshow(grid)
plt.colorbar()
```
