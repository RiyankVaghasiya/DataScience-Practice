```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y=x*x
```

```python
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0,6)  # we can set limit of axis
plt.ylim(0,15)
plt.title('This is label') 
plt.savefig('myfirstplot.png')  #this fun can save below image in your given location
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y=2*x
```

```python
fig = plt.figure(dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)

axes1 = fig.add_axes([0.1,0.7,0.25,0.25])
```

```python
# #creating subplot
fig = plt.figure()
axes = fig.subplots(nrows = 2, ncols = 2)
axes[0][1].plot(x,y)
axes[1][1].plot(a,b)
plt.tight_layout()  #this fun used for making space between graphs
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label = 'X vs X')
ax.plot(x[0:5],x[0:5]**2 , 'r',label = 'X vs X^2')
ax.plot(x[4:],x[4:]**2 ,'r--')


ax.legend(loc = (1.02,0.5))
```

__Bar chart__

```python
labels = ['A','B','C']
values = [1,2,3]
plt.bar(labels,values)
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"] 
num_oscars = [5, 11, 3, 8, 10]
```

```python
plt.bar(movies,num_oscars)
plt.title("movies list")
plt.xlabel("movies name")
plt.ylabel("number of oscars")
plt.show()
```

```python
years = [2017, 2018] 
mentions = [500, 505] 
plt.bar(years, mentions, 0.8) 
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")
plt.xlabel("years")
plt.show()
```
