```python
import pandas as pd 
```

```python
bios_data = [
   { "id": "1",
    "name":"Riyank",
    "born-date":"15 july",
    "born-country":"IND"},
    
    { "id": "2",
    "name":"Ruchit",
    "born-date":"5 jan",
    "born-country":"AUD"},

    { "id": "3",
    "name":"Drashti",
    "born-date":"15 jan",
    "born-country":"USA"},
    
    { "id": "4",
    "name":"Hitu",
    "born-date":"8 oct",
    "born-country":"GER"},

    { "id": "5",
    "name":"Meet",
    "born-date":"21 june",
    "born-country":"CND"},
    
]
```

```python
bios = pd.DataFrame(bios_data)
```

```python
bios
```

```python
noc_data = [
    {"noc":"IND",
    "region":"india"},
    
    {"noc":"CND",
    "region":"canada"},

    {"noc":"USA",
    "region":"america"},

    {"noc":"AUD",
    "region":"australia"},

    {"noc":"GER",
    "region":"germany"}
]
```

```python
nocs = pd.DataFrame(noc_data)
```

```python
nocs
```

```python
new_bios = pd.merge(bios,nocs,left_on="born-country",right_on = "noc",how = "left")
```

```python
new_bios = new_bios.drop(columns=['noc'])
```

```python
new_bios
```

```python
concat_df = pd.concat([bios,nocs])
```

```python
concat_df
```

```python
concat_df.dropna()
```
