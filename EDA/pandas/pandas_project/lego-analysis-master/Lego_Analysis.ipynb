```python
import pandas as pd
```

```python
df = pd.read_csv('lego-analysis-master/datasets/lego_sets.csv')
theme = pd.read_csv('lego-analysis-master/datasets/parent_themes.csv')
theme.head()
```

__Question:1 what percentage of all licensed sets ever released were star wars themed?__

```python
merged = df.merge(theme, left_on='parent_theme',right_on='name')
```

```python
merged = merged.drop(columns='name_y')
```

```python
merged.head()
```

```python
result = merged[(merged['parent_theme'] == 'Star Wars') & (merged['is_licensed'] == True)]

merged = merged.dropna(subset=['set_num'])

only_lincensed = merged[merged['is_licensed'] == True]

the_force = int((result.shape[0]/only_lincensed.shape[0])*100)
```

```python
the_force
```

__Question:2 in which year star wars not the most popular lincesed theme(in terms of number of sets released that year)?__

```python
sorted = only_lincensed.sort_values('year')
sorted['count'] = 1
grouped = sorted.groupby(['year','parent_theme']).sum().reset_index()
result_df = grouped.sort_values('count',ascending=False).drop_duplicates('year')
result_df.sort_values('year',inplace=True)
result_df
```
