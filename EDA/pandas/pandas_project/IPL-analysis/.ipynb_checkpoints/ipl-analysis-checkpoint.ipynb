```python
#make function for this type of task
def scored_most_runs(batsman):
    mask = df['batsman'] == batsman
    fliter_data = df[mask]
    return fliter_data.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
scored_most_runs('RG Sharma').head()
```

```python
# findout most destructive death over batsman in the history of the ipl
# on the bases of strike rate
# minimun playes 200 balls in 16-20 overs
```

```python
#first findout dataframe in which all the batsman are who play in 16-20 over
mask = df['over'] >= 16
death_over = df[mask]

#now second step is to remove batsman who faced less then 200 balls
all_batsman = death_over.groupby('batsman')['batsman_runs'].count() #this will give number of deliveries batsman has faced
mask2 = all_batsman > 200
filtered_batsman_list = all_batsman[mask2].index.tolist()  #it will take name only of batsman in index and convert it into series
#filtered_batsman_list

#now third step is to find strike rate of batsman
#formula = run scored/balls * 100
final_batsman = death_over[death_over['batsman'].isin(filtered_batsman_list)]
runs = final_batsman.groupby('batsman')['batsman_runs'].sum()
balls = final_batsman.groupby('batsman')['ball'].count()
final_ans = (runs /balls) * 100
x = final_ans.sort_values(ascending = False).head()
```

```python
x.plot(kind = 'barh',xlabel="strike rate")
plt.show()
```
