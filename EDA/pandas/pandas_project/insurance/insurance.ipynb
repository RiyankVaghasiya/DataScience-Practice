```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('insurance.csv')
```

## task1:-clean the data

__- find missing value and drop it__

```python
cleaned_data = data.dropna()
```

```python
cleaned_data.info()
```

__- clean region column__

```python
cleaned_data = cleaned_data.copy()
cleaned_data['region'] = cleaned_data['region'].str.lower()
```

```python
cleaned_data['region'].unique()
```

__- clean charges column__

```python
cleaned_data['charges'] = cleaned_data['charges'].str.strip('$').astype(dtype='float64')
```

__- clean the sex column__

```python
cleaned_data['sex'].unique()
cleaned_data['sex'] = cleaned_data['sex'].str.lower()
```

```python
 cleaned_data['sex'] = cleaned_data['sex'].replace({'woman':'female', 'f':'female', 'man':'male', 'm':'male'})
```

```python
cleaned_data['sex'].unique()
```

```python
#changing children column datatype
#changing smoker column datatype
cleaned_data['children'] = cleaned_data['children'].astype(dtype='int')
cleaned_data['smoker'] = cleaned_data['smoker'].replace({'yes':1 , 'no':0})
```

```python
cleaned_data.info()
```

```python
cleaned_data['age'] = cleaned_data['age'].abs()
```

```python
plt.scatter(cleaned_data['age'],cleaned_data['charges'])
plt.show()
```
