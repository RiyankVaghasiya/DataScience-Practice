```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
```

```python
train = pd.read_excel('Data_Train.xlsx')
```

```python
train.head()
```

```python
test = pd.read_excel('Test_set.xlsx')
```

```python
test.head()
```

```python
df = pd.concat([train,test])
```

```python
df.head()
```

```python
df.info()
```

__convert date of journey to three different columns__

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
#convert info into int
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year']= df['Year'].astype(int)
```

```python
df.info()
```

__convert total stops into numerical__

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
df['Total_Stops'] = df['Total_Stops'].fillna(1)
```

```python
df['Total_Stops'] = df['Total_Stops'].astype(int)
```

```python
df.info()
```

__changes in arrival time and departure time__

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minutes'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.drop('Arrival_Time',inplace=True,axis = 1)
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minutes'] = df['Arrival_minutes'].astype(int)
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_minutes'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.drop('Dep_Time',inplace=True,axis = 1)
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_minutes'] = df['Dep_minutes'].astype(int)
```

```python
df.head(2)
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.info()
```

__convert duration into minutes__

```python
df.head()
```

```python
df.replace(to_replace="5m",value="0.08h",inplace=True)
```

```python
df['duration_hour'] = df['Duration'].str.split('h').str[0]
df['duration_minutes'] = df['Duration'].str.split('h').str[1]
```

```python
df['duration_minutes'] = df['duration_minutes'].str.split('m').str[0]
```

```python
df['duration_minutes'].replace(to_replace='',value='0',inplace=True)
```

```python
df['duration_hour'] = df['duration_hour'].astype(float)
df['duration_minutes'] = df['duration_minutes'].astype(int)
```

```python
df.head()
```

```python
df['duration_hour'] = df['duration_hour'] * 60
```

```python
df['Duration'] = df['duration_hour'] + df['duration_minutes']
```

```python
df.drop(['duration_hour','duration_minutes'],axis=1,inplace=True)
```

```python
df['Duration'].replace(to_replace=4.8,value=5,inplace=True)
```

```python
df['Duration'] = df['Duration'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Airline'].unique()
```

__now we have to use machine learning algorithm like label encoding and onehot encoding to convert multiple categorical data into numerical data__

```python
# arr = df['Airline'].values
```

```python
encode = OneHotEncoder(sparse=False)
encoded_data = encode.fit_transform(df[['Airline']])
```

```python
encoded_df = pd.DataFrame(encoded_data, columns=encode.get_feature_names_out(['Airline']))
```

```python
encoded_df.head()
```

```python
df = df.reset_index(drop=True)
encoded_df = encoded_df.reset_index(drop=True)

final_df = pd.concat([df, encoded_df], axis=1)
```

```python
final_df.drop(columns=['Airline','Additional_Info'],inplace = True)
```

```python
final_df.head()
```

```python
final_df.columns
```

__convert source and destination columns into numerical__

```python
encode_source= OneHotEncoder(sparse = False)
encode_source_array = encode_source.fit_transform(df[['Source']])
```

```python
encoded_source_df = pd.DataFrame(encode_source_array, columns=encode_source.get_feature_names_out(['Source']))
```

```python
encode_destination = OneHotEncoder(sparse=False)
encode_destination_array = encode_destination.fit_transform(df[['Destination']])
```

```python
encoded_destination_df = pd.DataFrame(encode_destination_array, columns=encode_destination.get_feature_names_out(['Destination']))
```

```python
final_df = pd.concat([final_df,encoded_source_df,encoded_destination_df],axis=1)
```

```python
final_df = final_df.drop(columns=['Source','Destination'],axis=1)
```

```python
final_df.head()
```

```python
final_df['Price'] = final_df['Price'].fillna(np.mean(final_df['Price']))
```

```python
final_df.head()
```

__now its ready to train model__

```python
final_df.info()
```

```python
demo = final_df['Price']
```

```python
final_df = final_df.astype(int)
```

```python
final_df.drop(columns='Price',inplace=True)
```

```python
final_df['Price'] = demo
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
X = final_df.drop(columns='Price',axis=1)
Y = final_df['Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
testing_prediction = lr.predict(x_test)
```

```python
r2 = r2_score(y_test, testing_prediction)
```

```python
r2
```

```python
mae = mean_absolute_error(y_test,testing_prediction)
```
